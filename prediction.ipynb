```python
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
```

```python
myModel=tf.keras.models.load_model('My_module.h5')
```

```python
import numpy as np


def predict_image(sample_image,CNN_model,image_size=None):
    predict=tf.keras.preprocessing.image.load_img(sample_image,target_size=image_size)
    predict=tf.keras.preprocessing.image.img_to_array(predict)
    predict=predict/255
    predict=np.expand_dims(predict,axis=0)
    result=CNN_model.predict(predict)
    y_pred=np.argmax(result,axis=1)

    plt.imshow(plt.imread(sample_image),cmap='gray');
    plt.axis('off');

    if y_pred==1:
        return 'It is a Dog'
    else:
        return 'It is a Cat'
```

```python
image='C:/Users/Hp/Nural network/image prossesing/Petimages/test_set/dogs/dog.4005.jpg'

predict_image(image, myModel, image_size=(64,64))
```
